```python
import pandas as pd
data=pd.read_excel('PV.xlsx')
t=list(data.A)
a1=list(data.B)
a2=list(data.C)
a3=list(data.D)
a4=list(data.E)
a5=list(data.F)
a6=list(data.G)
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rc('xtick', labelsize=25)
matplotlib.rc('ytick', labelsize=25)
plt.figure(figsize=(20,12))
plt.plot(t, a1, 'r-', linewidth=1)
plt.plot(t, a2, 'y*', linewidth=1)
plt.plot(t, a3, 'g-', linewidth=1)
plt.plot(t, a4, 'y-', linewidth=1)
plt.plot(t, a5, 'p-', linewidth=1)
plt.plot(t, a6, 'b-', linewidth=1)
plt.minorticks_on()
plt.grid(True, which="both")
plt.axis([0,5250, 0.4,1.2])
plt.legend(['Shiroro',
            'Kano',
            'Gombe',
            'Jos',
            'Kaduna',
            'Katampe'], fontsize="xx-large")
plt.xlabel('Power [MW]', size=25)
plt.ylabel('Voltage in p.u', size=25)

```
